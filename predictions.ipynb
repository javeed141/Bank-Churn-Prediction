```python
import tensorflow as tf 
from tensorflow.keras.models import load_model
import pandas as pd 
import numpy as np
```

```python
import pickle
model=load_model("model.h5")

with open("label_encoder_gender.pkl","rb") as file :
    label_encoder_gender=pickle.load(file)


with open("onehot_encoder_geo.pkl","rb") as file :
    label_encoder_geo=pickle.load(file)
label_encoder_geo.get_feature_names_out(["Geography"])
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
data=pd.DataFrame([input_data])
```

```python
data
```

```python
# One-hot encode 'Geography'
geo_encoded = label_encoder_geo.transform(data[["Geography"]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=label_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data=pd.concat([data.drop("Geography",axis=1),geo_encoded_df],axis=1)
```

```python
data
```

```python
data["Gender"]=label_encoder_gender.transform(data["Gender"])
```

```python
data
```

```python
with open("scaler.pkl","rb") as file :
    scaler=pickle.load(file)
scaler
```

```python
input_scaled=scaler.transform(data)
input_scaled
```

```python
prediction=model.predict(input_scaled)
```

```python
prediction
```

```python
pred=prediction[0][0]
```

```python
if pred > 0.5:
    print("leaves bank")
else:
     print("dooesn't leaves bank")
```
